## cv2.cornerSubPix()

```python
import numpy as np
import cv2 as cv

if __name__ == "__main__":
    # termination criteria
    criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)

    # prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
    objp = np.zeros((6*9,3), np.float32)
    objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

    # Arrays to store object points and image points from all the images.
    objpoints = [] # 3d point in real world space
    imgpoints = [] # 2d points in image plane.
    
    # without refining
    img = cv.imread("C:\\Users\\Autonav\\github_repo\\ComputerVision\\BasicExercise_py\\chessboard3.jpg")
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    # Find the chess board corners
    ret, corners = cv.findChessboardCorners(gray, (9,6), None)

    # Draw and display the corners
    cv.drawChessboardCorners(img, (9,6), corners, ret)
    cv.imshow('img', img)

    # with refining

    # If found, add object points, image points (after refining them)
    if ret == True:
        objpoints.append(objp)
        corners2 = cv.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)
        imgpoints.append(corners2)

    # Draw and display the corners
    cv.drawChessboardCorners(img, (9,6), corners2, ret)
    cv.imshow('img2', img)
    
    cv.waitKey()
    cv.destroyAllWindows()
```

## cv2.drawChessoardCorners()

cv2.drawChessoardCorners(image, patternSize, corners, patternWasFound)

image - 컬러 이미지

patternSize - 체스보드 내부에 있는 코너의 갯수를 (행,열)로 나타낸다.

corners - cv2.findChessboardCorners 찾았거나 cv2.cornerSubPixel로 정제한 찾은 corner 좌표들이 담긴 자료형을 넣는다.

patternWasFound - cv2.findChessboardCorners로 코너를 찾았는지 못찾았는지 반환되서 나온 ret 값을 넣는다.
